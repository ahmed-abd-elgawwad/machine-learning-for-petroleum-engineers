```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
x= np.linspace(0,10,100)
y=x**3
```

```python
x
```

```python
y
```

```python
# fig=plt.figure(figsize=(5,5))
plt.plot(x,y,'red', linewidth=4, linestyle="--")
plt.xlabel('Time')
plt.ylabel('Number of iterations')
plt.title('Number of Iterations Vs. Time')
```

```python
# 1 represents row, 3 represents columns, the last number represnts the order. 
fig=plt.figure(figsize=(10,5))
plt.subplot(1,3,1)
plt.plot(x,y,'b')
plt.title('Plot 1')
plt.subplot(1,3,2)
plt.plot(y*2,x**2,'r')
plt.title('Plot 2')
plt.subplot(1,3,3)
plt.plot(x,y,'purple')
plt.title('Plot 3')
```

```python
fig= plt.figure()
axes= fig.add_axes([1,1,1,1])
axes.plot(x,y)
axes.set_xlabel('Time')
axes.set_ylabel('Number of iterations')
axes.set_title('Number of iterations vs. Time')
```

```python

# Plt.figure creates an empty figure
fig= plt.figure()
fig=plt.figure(figsize=(10,6))
# We then add the dimensions (left, bottom, width, height) for each figure

fig1= fig.add_axes([0.1,0.1,0.8,0.8])
fig2= fig.add_axes([0.2,0.6,0.3,0.2])

# We then add the axes and title for each figure
fig1.plot(x,y)
fig1.set_title('Large Plot')
fig1.set_xlabel('Time')
fig1.set_ylabel('Number of iteration')

fig2.plot(x*2,y)
fig2.set_title('Small Plot')
fig2.set_xlabel('Time')
fig2.set_ylabel('Number of iteration')
```

```python
fig,axes= plt.subplots(nrows=2, ncols=2,figsize=(5,5),dpi=100)

# axes.plot(x,y)

# Plot #1 in the first row and first column
axes[0,0].plot(x,y)
axes[0,0].set_title('Plot 1')
axes[0,0].set_xlabel('Time')
axes[0,0].set_ylabel('Production 1')

# Plot #2 in the first row and second column
axes[0,1].plot(y,x)
axes[0,1].set_title('Plot 2')
axes[0,1].set_xlabel('Time')
axes[0,1].set_ylabel('Production 2')

# Plot #3 in the second row and first column
axes[1,0].plot(y,x**2)
axes[1,0].set_title('Plot 3')
axes[1,0].set_xlabel('Time')
axes[1,0].set_ylabel('Production 3')

# Plot #4 in the second row and second column
axes[1,1].plot(x,y**2)
axes[1,1].set_title('Plot 4')
axes[1,1].set_xlabel('Time')
axes[1,1].set_ylabel('Production 4')

plt.tight_layout()

```

```python
g= np.linspace(1,10,5)
h= g**2
```

```python
fig=plt.figure(figsize=(10,10))
ax= fig.add_axes([0,0,1,1])


# RGB Hex codes are used for custom colors (https://www.rapidtables.com/web/color/RGB_Color.html)
# Line styles such as '--', '-.', 'steps' can be used
# marker such as 'o' and '+' can be used to show markers on a plot

ax.plot(g,h, color='#FF8C00', linewidth=3, linestyle='-.', alpha= 1, marker='o', markersize=20, markerfacecolor='green',
       markeredgewidth=3, markeredgecolor='red') 

ax.set_xlabel('Time')
ax.set_ylabel('Pricing')
ax.set_title('Time Vs. Pricing')

ax.set_xlim([0,8])
ax.set_ylim([0,80])


plt.rc('font', size=18)          # Controls default text sizes
plt.rc('axes', titlesize=18)     # Fontsize of the axes title
plt.rc('axes', labelsize=18)     # Fontsize of the x and y labels
plt.rc('xtick', labelsize=18)    # Fontsize of the tick labels
plt.rc('ytick', labelsize=18)    # Fontsize of the tick labels
plt.rc('legend', fontsize=18)    # Legend fontsize
plt.rc('figure', titlesize=18)   # Fontsize of the figure title

ax.text(3,3, "Commodity Pricing Plot", fontsize=30, color="red")
```
