```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
df_Shale= pd.read_csv('Chapter2_Shale_Gas_Wells_DataSet.csv')
```

```python
df_Shale.columns
```

```python
fig=plt.figure(figsize=(5,5))
sns.distplot(df_Shale['Well Spacing'], color='blue', bins=None, kde=False)
plt.xlabel('Well Spacing')
plt.ylabel('Frequency')
plt.title('Distribution Plot for Well Spacing')
```

```python
from scipy import stats
sns.jointplot(df_Shale['Stage Spacing'], df_Shale['EUR'], color='blue',kind='scatter',stat_func=stats.pearsonr )
```

```python
from scipy import stats
sns.jointplot(df_Shale['Proppant Loading'], df_Shale['EUR'],color='blue',kind='kde',stat_func=stats.pearsonr )
```

```python
sns.pairplot(df_Shale,hue='Category', kind='scatter', diag_kws={'bw': 1.5})
```

```python
 sns.lmplot(x='Lateral Length',y='EUR', data=df_Shale, hue='Category')
```

```python
sns.lmplot(x='Porosity',y='Water Saturation',data=df_Shale,col='Category',aspect=0.8,height=5)
```

```python
sns.barplot(df_Shale['Category'],df_Shale['EUR'],estimator=np.mean)
```

```python
sns.countplot(df_Shale['Category'])
```

```python
sns.boxplot(df_Shale['Porosity'], color='green')
```

```python
sns.boxplot(df_Shale['Category'],df_Shale['EUR'],palette='rainbow')
```

```python
sns.violinplot(df_Shale['Category'],df_Shale['Well Spacing'], palette='rainbow')
sns.swarmplot(df_Shale['Category'],df_Shale['Well Spacing'], palette='rainbow')
```

```python
sns.violinplot(df_Shale['Category'],df_Shale['Well Spacing'], palette='rainbow')
sns.boxplot(df_Shale['Category'],df_Shale['Well Spacing'], palette='coolwarm')
```

```python
sns.swarmplot(df_Shale['Category'],df_Shale['Well Spacing'], palette='rainbow')
sns.boxplot(df_Shale['Category'],df_Shale['Well Spacing'], palette='coolwarm')
```

```python
fig=plt.figure(figsize=(6,6))
sns.violinplot(df_Shale['Category'],df_Shale['Well Spacing'], palette='tab20')
```

```python
sns.kdeplot(df_Shale['Water Saturation'],df_Shale['EUR'],cmap='Reds', shade=True, shade_lowest=False)
```

```python
fig=plt.figure(figsize=(17,8))
sns.heatmap(df_Shale.corr(), cmap='coolwarm', annot=True, linewidths=4, linecolor='black')
```
