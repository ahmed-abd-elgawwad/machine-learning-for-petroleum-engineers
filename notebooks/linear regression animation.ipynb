```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from sklearn.linear_model import LinearRegression
from IPython import display


# Generate some random data
np.random.seed(42)
x = np.random.rand(100)
y = 2 * x + 1 + 0.1 * np.random.randn(100)

# Create a linear regression model
model = LinearRegression()

# Define the figure and axis for the plot
fig, ax = plt.subplots()

# Define the x and y limits of the plot
ax.set_xlim(0, 1)
ax.set_ylim(0, 4)

# Plot the initial data points
ax.scatter(x, y)

# Define the line object for the linear regression line
line, = ax.plot([], [])

# Define the function that will be called on each frame of the animation
def animate(i):
    # Fit the linear regression model on a subset of the data
    model.fit(x[:i+1].reshape(-1, 1), y[:i+1])

    # Generate the predictions for the entire range of x values
    y_pred = model.predict(np.linspace(0, 1, 100).reshape(-1, 1))

    # Update the line object with the new predictions
    line.set_data(np.linspace(0, 1, 100), y_pred)

    # Return the line object for the current frame
    return line,

# Define the animation object
animation = FuncAnimation(fig, animate, frames=len(x), interval=100, blit=True)

video = animation.to_html5_video()
html = display.HTML(video)
display.display(html)
# Show the animation
plt.close()
```
