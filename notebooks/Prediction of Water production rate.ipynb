## <div style="background-color:#F1C40F;padding:15px;border-radius:10px;text-align:center">Building the Models </div>

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# create time feaure
df["days"] = (df["DATEPRD"] - df["DATEPRD"].min()).dt.days
```

```python
X = df[["days",'AVG_DOWNHOLE_PRESSURE',"AVG_DOWNHOLE_TEMPERATURE",'AVG_DP_TUBING'
     ,'AVG_WHP_P','AVG_WHT_P','DP_CHOKE_SIZE']]
# X =  df[["days",'AVG_DP_TUBING','DP_CHOKE_SIZE']]
y = df['water_rate']
```

```python
# splitting the data to frecast 
train_size=  int(.8 * len(X))
X_train = X.iloc[:train_size]
y_train = y.iloc[:train_size]
X_test = X.iloc[train_size:]
y_test = y.iloc[train_size:]
```

```python
# # tansform the data
# X_scalar = MinMaxScaler()
# y_scalar = MinMaxScaler()
# X_scalar.fit(X_train)
# y_scalar.fit(np.array(y_train).reshape(-1,1))
# X_train_scaled = X_scalar.transform(X_train)
# X_test_scaled = X_scalar.transform(X_test)
# y_train_scaled = y_scalar.transform(np.array(y_train).reshape(-1,1))
# y_test_scaled = y_scalar.transform(np.array(y_test).reshape(-1,1))
```

```python
plt.title("Splitting the data For Frecasting in the future",fontsize=20,pad=50)
plt.plot(X_train["days"],y_train,color="blue")
plt.plot(X_test["days"],y_test,color="blue")
plt.xlabel("Time (days)")
plt.ylabel("Production Rate")
plt.axvline(x = X["days"].iloc[train_size], color = '#fff', label = 'axvline - full height',linewidth=5);
plt.annotate('Train-Data', xy=(600, 250), xytext=(600, 250),fontsize=18)
plt.annotate('Test-Data', xy=(600, 250),  xytext=(1100, 250),fontsize=18);
 
```

```python
from sklearn.linear_model import LinearRegression
# caclulate mean absolute error in each
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
lr = LinearRegression(normalize=False)
lr.fit(X_train,y_train)
```

```python
## make a function to show the model evaluationis

# get the mean absolute error
y_pred = lr.predict(X_test)
y_train_pred = lr.predict(X_train)
# inverse transform the y_pred
# y_pred = y_scalar.inverse_transform(np.array(y_pred).reshape(-1,1))
testing_mae = mean_absolute_error(y_test,y_pred)
testing_mse = mean_squared_error(y_test,y_pred)
training_mae = mean_absolute_error(y_train,y_train_pred)
training_mse =mean_squared_error(y_train,y_train_pred)
print(f"For Training  \n MAE :{training_mae} , MSE :{training_mse}")
print(f"For Testing  \n MAE :{testing_mae} , MSE :{testing_mse}")

# show the result
plt.figure(figsize=(10,6))
plt.title("Show model results",fontsize=20,pad=40)
plt.plot(X_test["days"],y_test,color="blue",label="True-Water")
plt.plot(X_test["days"],y_pred,color="#40E0D0",label="water_prediction")

plt.plot(X_train["days"],y_train,color="blue")
plt.axvline(x = X["days"].iloc[train_size], color = '#fff',linewidth=3, label = 'Begining of Prediciton');
plt.legend();
```
